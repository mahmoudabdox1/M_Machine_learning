```python
cleand_cars.drop("car_name",axis="columns",inplace=True)
```

```python
X = cleand_cars[['cyl', 'disp', 'hp', 'wt', 'acc', 'yr', 'origin', 'car_type']]
```

```python
X = cleand_cars.iloc[:,1:]
```

```python
X.columns
```

```python
y =  cleand_cars['mpg']
```

```python
y= cleand_cars.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=7)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import tensorflow as tf
print(tf.__version__)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

# Initialize the constructor
model = Sequential()
```

```python
ncols = X_train.shape[1]
ncols
```

```python
model = Sequential()
```

```python
model.add(Dense(2048,input_shape=(ncols,),activation="relu"))
model.add(Dense(1024,activation="relu"))
# Add an output layer with one neuron and no activation specified for regression
model.add(Dense(1))
```

```python

model.summary()
```

```python
optimizer = tf.keras.optimizers.Adam() #learning Rate = 0.01
model.compile(loss='mean_squared_error',optimizer=optimizer , metrics = ['mae'])
```

```python
epochs = 150
history = model.fit(X_train, y_train, epochs=epochs, validation_split = 0.2, verbose =True )
hist  = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
```

```python
hist.head()
```

```python
# hist['']
```

```python
import matplotlib.pyplot as plt

plt.plot(hist['mae'])
plt.plot(hist['val_mae'])
plt.legend(("train" , "valid") , loc =0)
```

```python
loss = model.evaluate(X_test, y_test, verbose=False)
loss
```

```python
y_pred = np.round(model.predict(X_test))
```
